**Percentage of male and female survived or not survibed**

```r
# install.packages("dplyr")
# library("dplyr")
# install.packages("stringr")
# library("stringr")
a <- df %>% filter(Survived == 1  & Sex == "female")
percentgae_female_survived <- (length(a$PassengerId)*100) / length(df$PassengerId)
# percentgae_female_survived

b <- df %>% filter(Survived == 1 & Sex == "male")
percentgae_male_survived <- (length(b$PassengerId)*100) / length(df$PassengerId)
# percentgae_male_survived

c <- df %>% filter(Survived == 0  & Sex == "female")
percentgae_female_notsurvived <- (length(c$PassengerId)*100) / length(df$PassengerId)
# percentgae_female_notsurvived

d <- df %>% filter(Survived == 0 & Sex == "male")
percentgae_male_notsurvived <- (length(d$PassengerId)*100) / length(df$PassengerId)
# percentgae_male_notsurvived

new_data <- data.frame(
  male_survived = percentgae_male_survived,
  female_survived = percentgae_female_survived,
  male_notsurvived = percentgae_male_notsurvived,
  female_notsurvived = percentgae_female_notsurvived
)
new_data
```

**count of male and female survived Not survived**

```r
library("dplyr")
a <- df %>% filter(Sex == "male" & Survived == 1)
new_data1 <- count(a)
b <- df %>% filter(Sex == "female" & Survived == 1)
new_data3 <- count(b)
c <- df %>% filter(Sex == "male" & Survived == 0)
new_data2 <- count(c)
d <- df %>% filter(Sex == "female" & Survived == 0)
new_data4 <- count(d)

data <- data.frame(
  count_male_survived = new_data1,
  count_female_survived = new_data2,
  count_male_notsurvived  =  new_data3,
  count_female_notsurvived = new_data4
)
colnames(data) <- c("count_male_survived" , "count_female_survived" , "count_male_notsurvived" , "count_female_notsurvived")
data
```
